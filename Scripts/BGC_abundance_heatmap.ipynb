```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/BiG-MAP.map.results.RPKM.csv")
df.head()
```

```python
# Aggregate the data: Sum abundance for each product prediction per sample
heatmap_data = df.groupby(["Sample", "Product Prediction"]).sum().reset_index()
```

```python
# Pivot the table to get samples as columns and product predictions as rows
heatmap_pivot = heatmap_data.pivot(index="Product Prediction", columns="Sample", values="Abundance")
```

```python
# Create the heatmap
plt.figure(figsize=(10, 8))  # Adjust figure size as needed
sns.heatmap(heatmap_pivot, annot=True, fmt=".2f", cmap="viridis", cbar_kws={"label": "Abundance"})

# Add titles and labels

plt.xlabel("Sample", fontsize=12)
plt.ylabel("BGC Product", fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```
