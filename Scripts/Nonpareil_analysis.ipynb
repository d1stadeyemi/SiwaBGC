```r
install.packages('Nonpareil');
```

```r
# Load Nonpareil library
library(Nonpareil);
```

```r
# Read the samples file
samples <- read.table('pareil.txt', sep='\t', header=TRUE, as.is=TRUE);
```

```r
print(samples)
```

```r
# Attach the columns for easier access
attach(samples);
```

```r
png("nonpareil_curves.png", width=800, height=600);
```

```r
# Generate and plot the Nonpareil curves
nps <- Nonpareil.set(File, col=Col, labels=Name, plot.opts=list(plot.observed=FALSE));
```

```r
# Close file
dev.off();
```

```r
# Detach the sample data
detach(samples);
```

```r
# View a summary of the Nonpareil results
summary(nps);
```
